```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page = requests.get('https://www.goodreads.com/quotes/tag/web')
```

```python
response = requests.get('https://www.goodreads.com/quotes/tag/web')
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
quote_entries = soup.find_all("div", class_="quoteDetails")
```

```python
quote_entries
```

```python
authors = []
texts = []
likes = []
for entry in quote_entries:
    author = entry.find("span", class_="authorOrTitle").text.strip()
    authors.append(author)
    
    text = entry.find("div", class_="quoteText").text.strip().split("\n")[0].strip("“”")
    texts.append(text)

    like_count = entry.find("a", class_="smallText").text.strip().split()[0]
    likes.append(like_count)
```

```python
data = {
    "Author": authors,
    "Text": texts,
    "Likes": likes
}
```

```python
df = pd.DataFrame(data)
```

```python
df.to_excel("goodreads_web_quotes.xlsx", index=False)
```
